# Análisis de los datos obtenidos de las simulaciones

```python
from matplotlib.pyplot import subplots
import pandas as pd
```

```python
datos = pd.read_csv("Resultados.csv")
datos
```

```python
datos = datos.set_index('Día')
```

```python
columnas_a_graficar = datos.columns  # todas las columnas
n_columnas = len(columnas_a_graficar)

# Crear subgráficas en disposición vertical
fig, axes = subplots(n_columnas, 1, figsize=(10, 4*n_columnas), sharex=True)

# Si solo hay una columna, axes no es una lista
if n_columnas == 1:
    axes = [axes]

for i, columna in enumerate(columnas_a_graficar):
    axes[i].plot(datos[columna], label=columna, color=f'C{i}')
    axes[i].set_ylabel(columna, fontsize=10)
    axes[i].legend(loc='upper right')
    axes[i].grid(True, alpha=0.3)
    
fig.suptitle('Variación diaria por activo', fontsize=14)
```
